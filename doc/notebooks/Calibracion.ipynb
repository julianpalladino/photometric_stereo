```python
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Cromada

```python
cromada = ndimage.imread('../res/ppmImagenes/cromada/cromada.0.ppm')
mask = ndimage.imread('../res/ppmImagenes/cromada/cromada.mask.ppm')
plt.imshow(cromada)
```

```python
#px, py = cx, cy
r = 119
cx, cy = (253, 147)
px, py = (286, 118) #.0

plt.figure(figsize=(16,16))
plt.imshow(mask)
plt.scatter(cx, cy, s=10, c='red', marker='o')
plt.scatter(px, py, s=10, c='blue', marker='o')
```

```python
nx = -(px - cx)
ny = py - cy
nx, ny
```

```python
nz = -np.sqrt(r*r - nx*nx - ny*ny)
nz
```

```python
n = np.array([nx, ny, nz]) / r
n
```

```python
L = -n
L
```

```python
from numpy.linalg import norm
norm(L)
```

```python
def light_direction(center, max_point):
    cx, cy = center
    px, py = max_point
    
    nx = -(px - cx)
    ny = py - cy
    nz = -np.sqrt(r*r - nx*nx - ny*ny)
    
    n = np.array([nx, ny, nz]) / r
    L = -n
    
    return L
```
